```python
#Finding schools with the best math scores
high_score = 800*0.8
#print(high_score)
math_high_scores = schools[schools['average_math']>=640]
#print(math_high_scores.head())
sorted_math_high_scores = math_high_scores.sort_values(by='average_math',ascending=False)
best_math_schools = sorted_math_high_scores[['school_name','average_math']]
print(best_math_schools)
```

```python
top_10_schools = schools.copy(deep=True)
top_10_schools['total_SAT'] = top_10_schools['average_math']+top_10_schools['average_reading']+top_10_schools['average_writing']
#print(top_10_schools['total_SAT']) 
top_10_schools = top_10_schools.sort_values(by='total_SAT',ascending=False)
top_10_schools = top_10_schools.loc[:,['school_name','total_SAT']]
top_10_schools = top_10_schools.head(10)
```

```python
schools_copy = schools.copy(deep=True)
schools_copy['total_SAT'] = schools_copy['average_math'] + schools_copy['average_reading'] + schools_copy['average_writing']
largest_std_dev = schools_copy.groupby('borough').agg(
    num_schools=('total_SAT', 'count'),
    average_SAT=('total_SAT', 'mean'),
    std_SAT=('total_SAT', 'std')
)
largest_std_dev = largest_std_dev.sort_values(by='std_SAT', ascending=False)
largest_std_dev['average_SAT'] = largest_std_dev['average_SAT'].round(2)
largest_std_dev['std_SAT'] = largest_std_dev['std_SAT'].round(2)
largest_std_dev = largest_std_dev.head(1)
```
